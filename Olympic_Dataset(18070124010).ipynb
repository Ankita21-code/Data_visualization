```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

Dataset Taken: Olympic Sports and Medals, 1896-2014 

```python
df=pd.read_csv('https://raw.githubusercontent.com/Ankita21-code/Data_visualization/main/summer.csv')
df.head()
```

#### BAR PLOT

Total Sports 

```python
ax = df['Sport'].value_counts().plot(kind='bar', color=['black','green'])
ax.set_xticklabels(['Aquatics','Athletics','Rowing',' Gymnastics', 'Fencing', 'Football', 'Hockey', 'Wrestling','Shooting', 'Sailing ', 'Cycling', 'Basketball', 'Canoe / Kayak', 'Volleyball', 'Handball ', 'Equestrian',
                    'Boxing', 'Weightlifting', 'Judo ','Baseball','Archery', 'Tennis ', 'Rugby', 'Modern Pentathlon', 'Softball','Table Tennis','Badminton','Taekwondo',
                    'Tug of War', 'Canoe', 'Polo', 'Lacrosse','Golf', 'Ice Hockey', 'Skating ', 'Cricket ', 'Triathlon', 'Rackets','Croquet', 'Water Motorsports',
                    'Basque Pelota ', 'Jeu de paume', 'Roque'])
ax.set(ylabel='Number of Passengers')
plt.show()
```

Gold Medalist

```python
ax = df[df['Medal']=='Gold']['Gender'].value_counts().plot(kind='barh', color=['black','green'])
ax.set_yticklabels(['Male','Female'])
ax.set(xlabel='Number of Medalists')
plt.show()
```

Gold, Silver and Bronze Medalist

```python
medalist_class=df.groupby(['Medal','Gender'])['Sport'].count().reset_index().sort_values(by='Gender',ascending=False)
medalist_class=medalist_class.pivot('Medal','Gender','Sport').fillna(0)
top=medalist_class
top.plot.barh(width=0.5,color=['blue','green'])
fig=plt.gcf()
fig.set_size_inches(8,8)
plt.title('Distribution Of the survived persons of the 3 medal classes')
plt.show()
```

#### SCATTER PLOT

```python
x=df['Year']
y=df['Discipline']

plt.xlabel("Year")
plt.ylabel("Discipline	")
plt.title("Scatter Graph")
plt.scatter(y,x,c="orange")
plt.legend()
plt.show()
```

#### PIE CHART

```python
s=df['Gender'].value_counts()
s
```

```python
num_males=s[0]
num_females=s[1]
```

```python
y = np.array([num_males,num_females])
mylabels = ["Males","Females"]
plt.pie(y,labels=mylabels)
plt.title('GENDER OF TOTAL ATHLETES')
plt.show()
```

```python
c=df['Medal'].value_counts()
c
```

```python
Bronze=c[0]
Silver=c[1]
Gold=c[2]
```

```python
y = np.array([Bronze,Silver,Gold])
mylabels = ["Bronze","Silver","Gold"]
plt.pie(y,labels=mylabels)
plt.title('MEDAL CLASSES')
plt.show()
```

#### LINE GRAPH

#### HEAT MAP 

```python
def olympics_corr(data):
    correlation = data.corr()
    sns.heatmap(correlation, annot=True, cbar=True, cmap="RdYlGn")
    
olympics_corr(df)
```

#### Heat Map of Null Values

```python
sns.heatmap(df.isnull(),yticklabels=False,cmap='viridis')
```
